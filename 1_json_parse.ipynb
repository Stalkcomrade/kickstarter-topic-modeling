```python
import pandas as pd

def CustomParser(data):
    import json
    j1 = json.loads(data)
    return j1

df = pd.read_csv('/home/stlk/Desktop/DigEc_data_samples/projects_json.csv', converters={'projects':CustomParser})

def check_keys(json, target):
    if target in json:
        return True
    for key in json:
        if isinstance(json[key], dict):
            if target in json[key]:
                return True
    return False

print(df.loc[df.projects.apply(check_keys, args=('urls',))])
```
