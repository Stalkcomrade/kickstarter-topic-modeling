```python
import pandas as pd
import numpy as np
```

```python
import rpy2.robjects as robjects
from rpy2.robjects.packages import importr
from rpy2.robjects import pandas2ri
pandas2ri.activate()


tidyr=importr('tidyverse')
magrittr=importr('magrittr')


rstring="""
    function(){  
    
    p2_log_clust = read_csv("~/share/research/ratings_univer/p2.log_clust.csv")

    # p2_log_clust
    p = p2_log_clust[,c(1:50, 65)] %>% as.data.frame()
    univer_names = p$univer

    p[p < -107] = 0
    p_w = p # weights

    # p[p > 0] = 1 # binary
    p = p[,-ncol(p)]

    p$univer = univer_names
    p = p %>% tidyr::gather(key = "company", value = "connection", 1:50)
    edgesWeight = p$connection

    # remove 0 edges

    p = p[p$connection != 0,]
    
    p
    
    }
"""

rfunc = robjects.r(rstring)
r_df = rfunc()
df = pandas2ri.ri2py(r_df)
```

```python

df = df.assign(un_id = range(0,df.shape[0]))
df.un_id += 1
```

```python
df2 = df.merge(df,on='company')

# df2
```

```python
df3 = df2[ df2.un_id_x > df2.un_id_y ]    # drop duplicates and matches to self
df3 = df3.assign(w = df3.connection_x * df3.connection_y)                  # multiply weights

# df3

```

```python
df4 = df3.groupby(['un_id_y','un_id_x'])['w'].sum()
df4


t = df4.to_frame()
t
```
